#Regla de aprendizaje de Hebb con decaimiento

```
import matplotlib.pyplot as plt
def hardlim(n):
  if (n<0): a=0
  else: a=1
  return a

w_vector = []
P0=1; W0=0 # Campana
P1=0; W1=1 #Alimento
b = -0.5
alpha = 0.6 #factor de aprendizaje
gamma = 0.3 #factor de olvido
#Razon aprendizaje = 2

for epoca in range(150):
  n = (P0*W0)+(P1*W1)+(b)
  a = hardlim(n) #tarjet
  #Regla de Hebb
  W0 = W0 + (alpha*a*P0) - (gamma*W0)
  if (epoca>10):
    P0=1; P1=1
  elif (epoca>100):
    P0=1; P1=0
  w_vector.append(W0)

plt.plot(w_vector,'*r')
```

#Regla Instar

```
import numpy as np
import math

def hardlim(n):
  if (n<0): a=0
  else: a=1
  return a

W=np.array([[ 12, 10, 25, 13, 27]]) #Patron Original

Pt=np.array([[15],[11],[25],[13],[33]]) #patron modificado

Wn = np.linalg.norm(W)
Pn = np.linalg.norm(Pt)

#theta=0
theta = (math.pi)/30 #6°

#Regla Instar
b = -(Wn*Pn*(math.cos(theta)))

n = np.dot(W,Pt)+b
a = hardlim(n)

print(f"a = {a}")
```
